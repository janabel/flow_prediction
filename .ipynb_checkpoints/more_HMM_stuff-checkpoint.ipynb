# Setting up architecture

We set up a Hidden Markov model with a multivariate Gaussian emission function. The latent variable here are the states of the underlying Markov chain, which could possibly represent incidents. Right now I don't expect the model to work super well, because in the HMM, the observed (emission) traffic flows do not depend on each other, only on the underlying state of traffic. Perhaps for n segments of roads, O(n!) states would be sufficient for capturing relative order of traffic flows. Perhaps Gaussian emissions also doesn't make much sense with a discrete set of hidden states. Thus we will look into coding up a Linear Dynamical System (LDS) to capture continuous hidden states (with linear evolution + small Gaussian noise), and ideally the traffic flow emissions from continuous hidden states makes more sense. 

One downside to encoding traffic states as hidden states is that the model is less interpretable - it is unclear what kinds of states these traffic states are referring to, and we need many in order to get a reasonable accuracy. If we include certain latent variables encoding incdients into the hidden states, again it's not clear what our underlying structure between incidents and non-incidents should be.

```python
# architecture parameters
num_components = 10 # sometimes does not converge when = 5
covar_type = "full"
```

```python
def vhmm_model(num_components, covar_type):
    converged = False
    while not converged:
        model = hmmlearn.vhmm.VariationalGaussianHMM(n_components=num_components, covariance_type=covar_type)
        model.fit(training_data, [training_index])
        converged = model.monitor_.converged
        if converged == False:
            print(f'model not converging for for num_components = {num_components}')
    return model
```

```python
# plot of log-likelihood probabilities on testing set

def plot_model_scores(a,b):
    models = [0] * b # array of models, indexed by the number of components
    scores = [0] * b
    for num_components in range(a, b):
        models[num_components] = vhmm_model(num_components, "full")
        scores[num_components] = models[num_components].score(testing_data, [n_samples-training_index])

    plt.figure(figsize=(20, 10), layout ='constrained')
    plt.plot(scores)
    plt.xlabel('num_components')
    plt.ylabel('log-likehood score of model')
    plt.title("Log-likehood for different number of components")
```

```python
plot_model_scores(5, 15)
```

```python
def train_all_models(columns, num_components_range, covar_type): 
    """columns - list, column names
    num_components_range - tuple, range for number of componenets in 
    covar_type - string, covariance type"""
    flow_data = raw_data[columns]
    flow_data_array = np.array(flow_data)
    
    # getting dimensions
    n_samples = flow_data_array.shape[0]
    n_features = flow_data_array.shape[1]

    # setting up training testing split
    training_index = math.floor(n_samples * 0.7)
    training_data = flow_data_array[0:training_index]
    testing_data = flow_data_array[training_index:]
    
    # initialize models and scores to be 0
    models = [0] * (num_components_range[1]-num_components_range[0]) # array of models, indexed by the number of components
    scores = [0] * (num_components_range[1]-num_components_range[0])
    
    # helper function for training each model
    def vhmm_model(num_components):
        converged = False
        while not converged:
            model = hmmlearn.vhmm.VariationalGaussianHMM(n_components=num_components, covariance_type=covar_type)
            model.fit(training_data, [training_index])
            converged = model.monitor_.converged
            if converged == False:
                print(f'model not converging for for num_components = {num_components}')
        return model
    
    # compute models and scores
    def compute_models():
        for num_components in range(num_components_range[0], num_components_range[1]):
            models[num_components-num_components_range[0]] = vhmm_model(num_components)
            scores[num_components-num_components_range[0]] = models[num_components-num_components_range[0]].score(testing_data, [n_samples-training_index])

    def plot_model_scores():
        plt.figure(figsize=(25, 15), layout ='constrained')
        plt.plot(scores)
        plt.xlabel('num_components')
        plt.ylabel('log-likehood score of model')
        plt.title("Log-likehood for different number of components")
        
    compute_models()
    plot_model_scores()
    return models, scores
```

```python
# testing out up to 24 hidden states - possibly all 4! = 24 permutations of the 4 travel segments
columns = ['400126','400176','400691','401242']
models, scores = train_all_models(columns, (10,25), "full")
```

```python
# plot of log-likelihood of trained model over number of components (# of hidden states in markov chain)
plt.figure(figsize=(25, 15), layout ='constrained')
plt.plot([i for i in range(10, 25)], scores)
plt.xlabel('num_components')
plt.ylabel('log-likehood score of model')
plt.title("Log-likehood for different number of components")
```

```python
best_model = models[-1]

means = best_model.means_ # mean of gaussian emission for each state (one vector / state)
covars = best_model.covars_
transmat = best_model.transmat_
```

```python
# need to compute most likely states for observed traffic flows for the first n timesteps
# then plot the means of those states as the predicted mean traffic flows
# plot against observed traffic flows for first n timesteps

col_strings_to_index = {'400126': 0,
                       '400176': 1, 
                       '400691': 2,
                       '401242': 3}

def predict_traffic_flows(model, column, end_time): # column = string, name of traffic segment
    means = best_model.means_
    covars = best_model.covars_
    transmat = best_model.transmat_
    
    col_index = col_strings_to_index[column]
    
    states = best_model.predict(training_data, [training_index])
    predicted_mean_traffic_flows = [means[s][col_index] for s in states]
    
    plt.figure(figsize=(25, 15), layout ='constrained')
    plt.plot(predicted_mean_traffic_flows, color='red', label='predicted traffic flow means')
    plt.plot(raw_data[column].tolist()[0:end_time], color='blue', label='actual traffic flows')
    plt.xlabel('time')
    plt.ylabel('traffic flow')
    plt.title("Traffic flow predictions on segments " + str(columns))
    plt.show()
```
