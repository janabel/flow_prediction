```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import biogeme.database as db
import biogeme.biogeme as bio
import biogeme.models as models
import biogeme.distributions as dist
import biogeme.results as res
import biogeme.messaging as msg
from biogeme.expressions import Beta, Variable, log

%config InlineBackend.figure_format='retina'
```

# Reading in the Data

```python
# read in travel time csv
time_df = pd.read_csv('Data/segment_travel_times.csv').dropna(axis=1).drop(columns=['Hour'])
time_df['Date'] = pd.to_datetime(time_df['Date'])
time_df.set_index('Date', inplace=True) # index travel times by date

# read in BART and driving demand csv's
bart_df = pd.read_csv('Data/bart_demands.csv')
driving_df = pd.read_csv('Data/driving_demands.csv')
bart_df['Date'] = pd.to_datetime(bart_df['Date'])
driving_df['Date'] = pd.to_datetime(driving_df['Date'])
bart_df.set_index('Date', inplace=True)
driving_df.set_index('Date', inplace=True)
demand_df = pd.concat([bart_df, driving_df], axis=1)

# create driving fraction clumn
demand_df['Driving Fraction'] = demand_df['Driving Flows'] / (demand_df['Driving Flows'] + \
                                                              demand_df['BART Flows'])
demand_df = demand_df[(demand_df['Driving Fraction'] < np.quantile(demand_df['Driving Fraction'],0.95)) & 
                     (demand_df['Driving Fraction'] > 0)] # remove extreme outliers

time_df = time_df.loc[demand_df.index]

incidents = pd.read_csv('Data/incidents_hour=7-8_year=2015-2019.csv')
incidents['Datetime'] = pd.to_datetime(incidents['3'])
incidents.set_index('Datetime', inplace=True)
ix = incidents.index
incident = [1 if x in pd.to_datetime(ix.date) else 0 for x in time_df.index]
```

# Data Exploration

## Histograms

```python
fig, axes = plt.subplots(3,5, figsize=(16,8))
years = list(range(2015, 2020))
for i in range(len(years)):
    temp = demand_df.loc[demand_df.index.isin(pd.bdate_range(f'{years[i]}-01-01', f'{years[i]}-12-31'))]
    axes[0,i].hist(temp['BART Flows'], bins=30)
    axes[0,i].set_xlim([demand_df['BART Flows'].min(), demand_df['BART Flows'].max()])
    axes[1,i].hist(temp['Driving Flows'], bins=30)
    axes[1,i].set_xlim([demand_df['Driving Flows'].min(), demand_df['Driving Flows'].max()])
    axes[2,i].hist(temp['Driving Fraction'], bins=30)
    axes[2,i].set_xlim([demand_df['Driving Fraction'].min(), demand_df['Driving Fraction'].max()])
    axes[2,i].set(xlabel=f'{years[i]}')
axes[0,0].set(ylabel='BART Flows')
axes[1,0].set(ylabel='Driving Flows')
axes[2,0].set(ylabel='Driving Fraction')
plt.show()
```

One interesting thing to note is the long-term shift in demand patterns. Why does it seem like demand is generally shifting in favor of the BART? Also, why do the BART flows start becoming bimodal around 2018? This may have to do with BART shutdowns or construction. Either way, it's worth looking into.

# Choice Model with Biogeme

```python
segments = time_df.columns
dataset = pd.concat([time_df]*2, axis=0)
dataset['Incident'] = incident * 2
dataset['Weight'] = demand_df['Driving Flows'].tolist() + demand_df['BART Flows'].tolist()
dataset['Weight'] /= (dataset['Weight'].sum() / len(dataset))
dataset['Choice'] = [1] * int(len(dataset)/2) + [0] * int(len(dataset)/2)
dataset.head()
```

```python
# create Biogeme database from dataframe
database = db.Database("SF", dataset)
globals().update(database.variables)

# create an attribute-specific constant for BART
ASC_BART = Beta('ASC_BART', 0, None, None, 0) # initialize Beta with value of 0 and no upper/lower bounds
V_BART = ASC_BART # define constant systematic utility for BART equal to ASC_BART

# initialize a systematic utility for Auto of 0 and add segment travel time variables in loop
V_Auto = 0
for segment in segments:
    globals()[f'B_{segment}'] = Beta(f'B_{segment}', 0, None, None, 0) # define segment variable coefficient
    V_Auto += globals()[f'B_{segment}'] * globals()[segment] # add to systematic utility

# create a dummy variable for incidents
B_Incident = Beta('B_INCIDENT', 0, None, None, 0)
V_Auto += B_Incident * Incident # add incident variable to Auto systematic utility

V = {0: V_BART, 1: V_Auto}
Logprob = log(models.logit(V, None, Choice))
Weight /= dataset['Weight'].max()

biogeme = bio.BIOGEME(database, {'loglike': Logprob, 'weight': Weight})
biogeme.modelName = 'logit'
biogeme.generateHtml = True
biogeme.generatePickle = False

# Define level of verbosity
logger = msg.bioMessage()
logger.setGeneral()

results = biogeme.estimate()
betas = results.getBetaValues()
```

```python
results.getEstimatedParameters()
```

The model results are saved as an HTML file and the estimated parameters can be viewed using the $\verb|getEstimtedParameters|$ method from Biogeme. We can see that the p-values are very high for all estimated coefficients. Moreover the HTML file tells us that the smallest eigenvalue is very close to zero. As expected, we have extreme multicollinearity in our dataset, which is causing difficulties with interpretation. Moreover, many of the estimated parameters are positive, which is counter-intuitive given that we would expect a negative effect of increased travel times on driving utility. Most likely, this is a case of endogeneity.
